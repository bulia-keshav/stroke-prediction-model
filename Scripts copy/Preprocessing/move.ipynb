```python
import os
import shutil
```

```python
def copy_files(src_dir, dest_dir):
    """
    Copy .nii and .nrrd files from source directory to destination directory.
    
    Args:
        src_dir (str): Path to the source directory.
        dest_dir (str): Path to the destination directory.
    """
    # Create destination directories if they don't exist
    if not os.path.exists(dest_dir):
        os.makedirs(dest_dir)
    
    # Iterate through folders in source directory
    for folder in os.listdir(src_dir):
        folder_path = os.path.join(src_dir, folder)
        if os.path.isdir(folder_path):
            # Iterate through files in each folder
            for file in os.listdir(folder_path):
                file_path = os.path.join(folder_path, file)
                if file.endswith('.nii') or file.endswith('.nrrd'):
                    dest_file_path = os.path.join(dest_dir, file)
                    # Copy file to destination directory
                    shutil.copy2(file_path, dest_file_path)
```

```python
import os

def count_files(directory):
    count = 0
    for _, _, files in os.walk(directory):
        count += len(files)
    return count

directory_path = 'C:/Users/Soham Sant/Desktop/Data Preparing/Masks'
file_count = count_files(directory_path)
print("Number of files:", file_count)
```

```python
import os

def find_duplicate_number(directory):
    expected_numbers = set(range(1, 131))  # Assuming files are named from 001 to 130
    actual_numbers = set()

    duplicate_numbers = set()

    for _, _, files in os.walk(directory):
        for file_name in files:
            if file_name.startswith("0"):  # To handle filenames like "001", "002", etc.
                try:
                    file_number = int(file_name[:3])
                except ValueError:
                    continue  # Skip files that don't follow the naming convention
            else:
                try:
                    file_number = int(file_name)
                except ValueError:
                    continue  # Skip files that don't follow the naming convention

            if file_number in actual_numbers:
                duplicate_numbers.add(file_number)
            else:
                actual_numbers.add(file_number)

    return duplicate_numbers

directory_path = 'C:/Users/Soham Sant/Desktop/Data Preparing/Masks'
duplicate_numbers = find_duplicate_number(directory_path)

if duplicate_numbers:
    print("Duplicate number(s) found:", duplicate_numbers)
else:
    print("No duplicate numbers found.")
```

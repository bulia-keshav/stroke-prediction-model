```python
import pandas as pd
```

```python
file_path = 'PhD MRI+ CTA+MRS 203+ 161.xlsx'  # Replace with your Excel file path
sheet_name = 'MRI+CTA+MRS 161'  # Replace with the actual name of the sheet you want to load

data = pd.read_excel(file_path, sheet_name=sheet_name)
```

```python
sheets = pd.ExcelFile(file_path).sheet_names
print(sheets)
```

```python
# Calculate the percentage of missing values for each column
missing_percentage = data.isna().mean() * 100
```

```python
# Create a DataFrame to display the results
missing_df = pd.DataFrame({
    'Column Name': missing_percentage.index,
    'Missing Percentage': missing_percentage.values
})

# Print the DataFrame
print(missing_df)
```

```python
# Optionally, save the results to a new Excel file
missing_df.to_excel('missing_values_percentage.xlsx', index=False)
```
